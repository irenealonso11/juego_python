```python
import random
```

```python

palabras = ["lupa", "vocal", "cuerda", "nube", "mural"]

palabra_secreta = random.choice(palabras)  # Escoger una palabra aleatoria

letras_adivinadas = []

for a in range(len(palabra_secreta)):
  letras_adivinadas.append('_')
  
letras_incorrectas = []
intentos = 6  # Número de intentos disponibles (de 0 a 6 fallos)



ahorcado_dibujos = [
    '''
      +---+
      |   |
          |
          |
          |
          |
    =========
    ''',
    '''
      +---+
      |   |
      O   |
          |
          |
          |
    =========
    ''',
    '''
      +---+
      |   |
      O   |
      |   |
          |
          |
    =========
    ''',
    '''
      +---+
      |   |
      O   |
     /|   |
          |
          |
    =========
    ''',
    '''
      +---+
      |   |
      O   |
     /|\\  |
          |
          |
    =========
    ''',
    '''
      +---+
      |   |
      O   |
     /|\\  |
     /    |
          |
    =========
    ''',
    '''
      +---+
      |   |
      O   |
     /|\\  |
     / \\  |
          |
    =========
    '''
]

print(f"La palabra 🔒 tiene {str(letras_adivinadas)} {len(palabra_secreta)} letras")
print("------------------------------------------")   

while intentos > 0:
        # Solicitar al usuario que ingrese una letra
        letra_usuario = input("Adivina una letra: ").lower()
        
        # Mostrar el dibujo actual del ahorcado
        print(ahorcado_dibujos[6 - intentos])
        
        # Mostrar el estado actual de la palabra
        posicion = palabra_secreta.find(letra_usuario.lower())
        print(f"Posicion letra es {posicion}")  
        letras_adivinadas[posicion] = letra_usuario
        
        # Mostrar las letras incorrectas
        print("Letras incorrectas:", " ".join(letras_incorrectas))
        
        # Verificar si la letra ya fue adivinada
        if letra_usuario in letras_adivinadas or letra_usuario in letras_incorrectas:
            print("Inténtalo otra vez...🤔")
            continue

        # Verificar si la letra está en la palabra secreta
        if letra_usuario in palabra_secreta:
            letras_adivinadas.append(letra_usuario)
            print(f"¡Bien hecho!👍 La letra '{letra_usuario}' está en la palabra!! 🎉🎉")
            
        else:
            letras_incorrectas.append(letra_usuario)
            intentos -= 1  # Restar un intento si la letra no está en la palabra
            print(f"Lo siento 🫣...la letra '{letra_usuario}' no está en la palabra...😥")
        
        # Verificar si el jugador ha ganado
        if all(letra in letras_adivinadas for letra in palabra_secreta):
            print("¡Felicidades! ¡Has ganado el juego!🥇🥇🥇 🎉🎉🎉")
            break
    
# Si el jugador pierde
if intentos == 0:
        print(ahorcado_dibujos[6])
        print(f"¡Lo siento! ¡Te has quedado sin intentos! La palabra era: {palabra_secreta}...🤷🏻‍♀️")


```

6 - intentos: Este cálculo sirve para seleccionar el dibujo correspondiente al número de intentos restantes. Si tienes 6 intentos en total, esta fórmula hace lo siguiente:

Cuando el jugador ha cometido 0 fallos, intentos será 6, y la expresión 6 - intentos será 0, mostrando el dibujo inicial (normalmente un poste vacío).
Cuando el jugador ha cometido 1 fallo, intentos será 5, por lo que 6 - intentos será 1, mostrando el siguiente dibujo del ahorcado (por ejemplo, la cabeza).
Y así sucesivamente hasta que los intentos lleguen a 0, momento en el cual se mostraría el dibujo final del ahorcado completo.


" ".join(letras_incorrectas):

El método join() se utiliza para concatenar todos los elementos de la lista letras_incorrectas en una sola cadena de texto, separándolos por un espacio " ".
Por ejemplo, si letras_incorrectas = ['b', 'z', 'q'], el resultado de " ".join(letras_incorrectas) será "b z q". Esto permite que las letras incorrectas se impriman de forma ordenada, separadas por espacios.
