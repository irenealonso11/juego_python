```python
import random
```

```python

palabras = ["lupa", "vocal", "cuerda", "nube", "mural"]

palabra_secreta = random.choice(palabras)  # Escoger una palabra aleatoria

letras_adivinadas = []
letras_incorrectas = []
intentos = 6  # Número de intentos disponibles (de 0 a 6 fallos)

long_palabra= len(palabra_secreta)
palabra_oculta= " _ " * long_palabra

ahorcado_dibujos = [
    '''
      +---+
      |   |
          |
          |
          |
          |
    =========
    ''',
    '''
      +---+
      |   |
      O   |
          |
          |
          |
    =========
    ''',
    '''
      +---+
      |   |
      O   |
      |   |
          |
          |
    =========
    ''',
    '''
      +---+
      |   |
      O   |
     /|   |
          |
          |
    =========
    ''',
    '''
      +---+
      |   |
      O   |
     /|\\  |
          |
          |
    =========
    ''',
    '''
      +---+
      |   |
      O   |
     /|\\  |
     /    |
          |
    =========
    ''',
    '''
      +---+
      |   |
      O   |
     /|\\  |
     / \\  |
          |
    =========
    '''
]

print(f"La palabra 🔒 tiene {palabra_oculta} {long_palabra} letras")
print("------------------------------------------")   

while intentos > 0:
        # Solicitar al usuario que ingrese una letra
        letra_usuario = input("Adivina una letra: ").lower()
        
        # Mostrar el dibujo actual del ahorcado
        print(ahorcado_dibujos[6 - intentos])
        
        # Mostrar el estado actual de la palabra
        posicion = palabra_secreta.find(letra_usuario.lower())
        
        # Mostrar las letras incorrectas
        print("Letras incorrectas:", " ".join(letras_incorrectas))
        
        # Verificar si la letra ya fue adivinada
        if letra_usuario in letras_adivinadas or letra_usuario in letras_incorrectas:
            print("Inténtalo otra vez...🤔")
            continue

        # Verificar si la letra está en la palabra secreta
        if letra_usuario in palabra_secreta:
            letras_adivinadas.append(letra_usuario)
            print(f"¡Bien hecho!👍 La letra '{letra_usuario}' está en la palabra!! 🎉🎉")
        else:
            letras_incorrectas.append(letra_usuario)
            intentos -= 1  # Restar un intento si la letra no está en la palabra
            print(f"Lo siento 🫣...la letra '{letra_usuario}' no está en la palabra...😥")
        
        # Verificar si el jugador ha ganado
        if all(letra in letras_adivinadas for letra in palabra_secreta):
            print("¡Felicidades! ¡Has ganado el juego!🥇🥇🥇 🎉🎉🎉")
            break
    
# Si el jugador pierde
if intentos == 0:
        print(ahorcado_dibujos[6])
        print(f"¡Lo siento! ¡Te has quedado sin intentos! La palabra era: {palabra_secreta}...🤷🏻‍♀️")


```
